#USING LIBRARY

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python

range_n_clusters = range(1, 21)

# Inisialisasi list untuk menyimpan nilai inersia
inertia_values = []

# Coba beberapa jumlah klaster
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(x)
    inertia_values.append(kmeans.inertia_)

# Plotting metode siku
plt.plot(range_n_clusters, inertia_values, marker='o')
plt.xlabel('Jumlah Klaster')
plt.ylabel('Inersia (Within-Cluster Sum of Squares)')
plt.title('Metode Siku untuk Menentukan Jumlah Klaster Optimal')
plt.show()
```

```python
range_n_clusters = range(2, 21)

# Inisialisasi list untuk menyimpan Silhouette Scores
silhouette_scores = []

# Coba beberapa jumlah klaster
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(x)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(x, labels)
    silhouette_scores.append(silhouette_avg)

# Plotting Silhouette Scores
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.xlabel('Jumlah Klaster')
plt.ylabel('Silhouette Score')
plt.title('Grafik Silhouette Score untuk Evaluasi Klastering')
plt.show()
```
